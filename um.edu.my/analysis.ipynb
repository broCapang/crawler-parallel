```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parallel execution times for 4 nodes (10 runs)
parallel_times = [
    [5.460802, 5.450911, 5.438591, 3.665400],
    [3.656574, 4.343122, 4.439711, 5.038829],
    [4.629778, 5.117870, 3.850603, 5.573647],
    [3.522676, 3.613763, 3.934131, 4.126039],
    [3.474368, 3.538594, 3.608517, 3.617689],
    [3.510679, 3.607758, 3.655292, 3.713603],
    [3.582148, 3.567154, 3.495260, 3.436936],
    [3.436107, 3.634895, 3.432074, 3.486762],
    [4.722977, 4.621980, 3.671696, 4.093771],
    [3.735996, 3.478786, 4.618665, 4.348304]
]

# Single node execution times (10 runs)
single_times = [
    10.246276, 10.259244, 10.262204, 10.035377, 10.461669, 
    10.223362, 10.961001, 10.104933, 11.346252, 11.396002
]

```

```python
parallel_averages = [np.mean(run) for run in parallel_times]
```

```python
# Calculate the overall average time for parallel and single node runs
overall_parallel_avg = np.mean(parallel_averages)
overall_single_avg = np.mean(single_times)
```

```python
# Plot the execution times
plt.figure(figsize=(12, 6))
# Plot parallel execution times
for i, run in enumerate(parallel_times):
    plt.plot(run, marker='o', linestyle='-', label=f'Parallel Run {i+1}')


# Highlight average lines
plt.axhline(y=overall_parallel_avg, color='red', linestyle='--', label='Avg Parallel Time')

# Add titles and labels
plt.title('Execution Times for Parallel Runs')
plt.xlabel('Run')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the execution times
plt.figure(figsize=(12, 6))

# Plot single node execution times
plt.plot(single_times, marker='x', linestyle='-', color='black', label='Single Node Runs')

# Highlight average lines
plt.axhline(y=overall_single_avg, color='blue', linestyle='--', label='Avg Single Node Time')

# Add titles and labels
plt.title('Execution Times for Single Node Runs')
plt.xlabel('Run')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the execution times
plt.figure(figsize=(12, 6))

# Highlight average lines
plt.axhline(y=overall_parallel_avg, color='red', linestyle='--', label='Avg Parallel Time')
plt.axhline(y=overall_single_avg, color='blue', linestyle='--', label='Avg Single Node Time')

# Add titles and labels
plt.title('AVERAGE Execution Times Comparison')
plt.xlabel('Run')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the average time for each parallel run
parallel_averages = [np.mean(run) for run in parallel_times]

# Calculate the overall average time for parallel and single node runs
overall_parallel_avg = np.mean(parallel_averages)
overall_single_avg = np.mean(single_times)

# Calculate the speedup
speedup = overall_single_avg / overall_parallel_avg
percentage_improvement = (1 - (overall_parallel_avg / overall_single_avg)) * 100

# Print the averages and speedup
print(f"Average execution time (parallel): {overall_parallel_avg:.6f} seconds")
print(f"Average execution time (single node): {overall_single_avg:.6f} seconds")
print(f"Speedup: {speedup:.2f}x")
print(f"Percentage improvement: {percentage_improvement:.2f}%")

# Plot the averages for comparison
plt.figure(figsize=(10, 6))
categories = ['Single Node', 'Parallel']
averages = [overall_single_avg, overall_parallel_avg]
plt.bar(categories, averages, color=['blue', 'green'])
plt.ylabel('Average Execution Time (seconds)')
plt.title('Comparison of Average Execution Times')
plt.show()
```
